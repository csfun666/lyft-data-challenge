```python
import pandas as pd
import numpy as np
```

```python
ride_ids = pd.read_csv("./data/ride_ids.csv")
driver_ids = pd.read_csv("./data/driver_ids.csv")
ride_timestamp = pd.read_csv("./data/ride_timestamps.csv")
```

```python
# price card
base_fare = 2
cost_per_min = 1.15
cost_per_mile = 0.22
service_fee = 1.75
minimum_fare = 5
maximum_fare = 400
```

```python
def calculate_fare(ride_duration, ride_distance, prime_time):
    # ride_distance: FLOAT, in miles
    # ride_duration: FLOAT, in minutes
    # prime_time: INTEGER, fare increase by this persent
    fare = (1 + prime_time / 100) *(base_fare + service_fee + cost_per_mile * ride_distance + cost_per_min * ride_duration)
    if fare < 5:
        fare = 5.00
    if fare > 400:
        fare = 400
    return fare
```

```python
fare_arr = []
for index, row in ride_ids.iterrows():
    distance = row[2] * 0.000621371
    time = row[3] / 60
    prime_time = row[4]
    fare = round(calculate_fare(time, distance, prime_time), 2)
    fare_arr.append(fare)
ride_ids["fare"] =  np.array(fare_arr)
ride_ids
```

```python
id_to_data = {}
for index, row in ride_timestamp.iterrows():
#     print(index)
    try:
        timestamp_day = row[2].split(" ")[0]
        if id_to_data.get(row[0], 0) == 0:
            id_to_data[row[0]] = timestamp_day
    except:
        print(row[0])
#     print(timestamp_day)
# id_to_data
```

```python
driver_with_day = {}
for index, row in ride_ids.iterrows():
    # driver_with_day["date_to_fare"]: a 4 elements array. The first is fare, 
    # the second is the distance of drving, the third is duration(time), 
    # and the forth is the number of ride
    
    driver_id = row[0]
    ride_id = row[1]
    distance = row[2]
    duration = row[3]
    fare = row[5]
    date = id_to_data.get(ride_id, -9999)
    
    if driver_with_day.get("id", -1) == -1:
        # the driver is not in the list, create data for the driver
        driver_with_day["id"] = driver_id
        driver_with_day["date_to_fare"] = {date:[fare, distance, duration, 1]}
    elif driver_with_day.get("id", -1) == driver_id:
        # the driver is in the list, keep update data
#         print(driver_with_day["date_to_fare"])
        date_result = driver_with_day["date_to_fare"].get(date, -1)
        if date_result == -1:
#             print(driver_with_day["date_to_fare"])
            driver_with_day["date_to_fare"][date] = [fare, distance, duration, 1]

        else:
            driver_with_day["date_to_fare"][date][0] += fare
            driver_with_day["date_to_fare"][date][1] += distance
            driver_with_day["date_to_fare"][date][2] += duration
            driver_with_day["date_to_fare"][date][3] += 1
    elif driver_with_day.get("id", -1) != driver_id:
        # next driver info, save previous driver info to csv
        date_val = []
        fare_val = []
        distance_val = []
        duration_val = []
        number_of_ride_val = []
#         print(driver_with_day["date_to_fare"])
        for key, value in driver_with_day["date_to_fare"].items():
            date_val.append(key)
            fare_val.append(value[0])
            distance_val.append(value[1])
            duration_val.append(value[2])
            number_of_ride_val.append(value[3])
        total = {"date":date_val, "fare":fare_val, "distance":distance_val, 
                 "duration":duration_val, "number of ride":number_of_ride_val}
#         print()
#         print(total)
        driver_data = pd.DataFrame(total)
        driver_data.to_csv("./driver/{0}.csv".format(driver_with_day["id"]))
        driver_with_day = {}
print("done")
```
